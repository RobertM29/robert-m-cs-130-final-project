## Why and How Do College Students Play Videogames 
         To begin, 54.9 % of survey respondents were in their 20s (7388), 42.7 % were 18 to 19 years old, and 52.5% of respondents (7073) were a student currently in a college or university. Thus, most respondents are students introduced into the realm of adult life and independence. So, they may turn to videogames for an ample amount of reasons. Well, what reasons in particular, one may ask? Well, based off data, most respondents say they play video games to have fun (39.2%) and improve (36.6%), while others focus on winning (15.5%), relaxing (4.9%), and a mix of all said so far (0.2%) along with other responses (3.6%). Thus, in the stresses of college life, students turn to video games to have an outlet that keeps their mind off of impending workloads that await them. They also aim to improve, which means they do not only play video games as an outside outlet, but they use them to learn from past mistakes. For example, if a college student plays Fortnite and uses the wrong weapons (let's say an SMG instead of a shotgun), then they know for next time not to use an SMG against a shotgun. Therefore, college students use videogames as a fun method to learn outside the classroom. Also, based on the whyplay data, college students likely use video games to destress and gain confidence from realizing they have the capability to win, which boosts their self-esteem, making them less unsure about their abilities and themselves. 
         
        To recap, college students play videogames to destress and gain confidence in their abilities, but how do they play these games? Well, gaming_anxiety receieved responses such as Multiplayer - online - with real life friends Percentage, Multiplayer - online - with strangers, Multiplayer - online - with aquaintances or teammates Percentage, Singleplayer Percentage, Multiplayer Offline Percentage, and Combined Playstyle Percentage. Among the data, most responded playing online multiplayer with real-life friends (41.3 %), playing online multiplayer with strangers (30.7%), and playing online multiplayer with aquiantances (19.7%). Now, some college students do play singleplayer (5.7%), multiplayer offline (0.4%), and some utilize every playstyle (2.3%).

        With the data whyplay and Playstyle in gaming_anxiety, one can come to the conslusion college students use videogames to balance their lives socially and recreationally. However, how does that hold true? 
        
print(![A bar chart relating to respondents' average hours gaming per week.](GamingHours.png)) Based off of data from gaming_anxiety, 44.8% of respondents play for less than 20 hours, while 75.8% of respondents play for less than 30 hours. Therefore, despite risks such as video game addiction, college students are able to balance their time. ## Are Video Games A Benefit or Detriment To College Students? 
        Answering the question asked earlier involves analyzing GADE (Affect of Gaming In Work), SWL (Life Satisfaction), and SPIN (Social Phobia Inventory). In the context of college students, GADE refers to how much of a distraction gaming is from academic performance or on-campus and off-campus jobs. SWL refers to how satisfied college students are with their lives on campus and at home when playing video games. And lastly, SPIN refers to how anxious college students are over challenging coursework, extracirriculars, and work in their schedules. In the context of video games, SPIN scores would detetrmine if video games truly help mitigate the stresses of college life. 
        Starting with GADE, survey respondents replied with "not difficult at all", "somewhat difficult", "very difficult", and "extremely difficult." Of these responses, 46.3% (6236) responded with "not difficult at all", while 38.2% (5148) responded with "somewhat difficult." Thus, through the lens of academics, one can assume focus and performance within college students benefits from video games, and that it may even improve over time depending on the game played. For example, Fortnite can increase reaction time through fast-paced gameplay, Roblox obbies can teach a person trial and error, and single-purchase story games can help a person pay attention of the plot. Therefore, different video games attest to different skills that help with academic focus and efficiency, which benefit the balance college students need to handle in their everyday lives. To further support this claim, only 7.5% of survey respondents (1012) replied with "very difficult", and 3.1% of survey respondents (419) replied with "extremely difficult." Hence, 84.5% of respodents playing video games have a benefit on their focus and efficiency, while 10.6% of respondents believe playing video games mainly hinders their focus and efficiency, while 4.9% have no response (out of the 13,464 responders, only 12,815 responded - based on math). Therefore, college students more effectively balance their time and academic performance using video games to exercise their brains. 
        Next, in SWL data, survey respondents replied with varying scores. 0.8% of respondents (114) replied with the highest score said (35), and 1.8% respondents (244) replied with the lowest score said (5). Then, 6.3 % of respondents (851) had a SWL score higher than 30, 47.3 % of respondents (6373) had a SWL score higher than 20, 48.2 % of respondents had a SWL score less than 20 (6484), and 8.6 % of respondents (1160) had a SWL score lower than 10. But what do these communicate about college students?

Life satisfaction for college likely comes from having a balanced life.         Lastly, similar to SWL data, survey respondents replied with a split in SPIN data. 51.2% of respondents (6887) replied with a SPIN Score less than 19, which means 48.8% of respondents (6577) responded a SPIN score more than 19. Therefore, just over half of respondents believe their anxiety is at a healthy level, while just under half believe it might be detrimental. Thus, for college students, it is probable adrenaline from video games might make them more nervous or anxious depending on the intensity of the game, and the amount of stress on their mind. However, college students might also find video games making them less nervous and anxious when they act as an "escape from reality" and stressors, even when EuStress remains. So, depending on which game works best for them, college students can diminish their anxiety, or they can make their anxiety worse. It depends on the game choice, and the data conveys there is not a clear "cut and dry" between whether video games increase or decrease anxiety. 
        However, only 0.5% of respondents (73) had a SPIN score higher than 60, which indicates extreme anxiety. Therefore, video games may create moderate nervousness within college students, but not enough to where it is detrimental to their overall health and well-being based on the data. So, video game choice is important to determine which game best suits a college student and their handling on the stresses and challenges they face in college life. 

## College Students Should Play Video Games - Here's Why 
        Ultimately, the data provided by gaming_anxiety shows video games bolster confidence, improve focus, and practice balance. However, the data also conveys the risks that come to how long one plays video games, and which video game one plays - addiction and high anxiety are two of these. Thus, college students should play video games to exercise their brains. However, similar to the muscles in our bodies, the muscles in our brains need rest as well by taking breaks via less brain-intensive activies: calling a friend, watching TV or YouTube videos, reading a book. Having our brains develop boredom at points does relax them, despite having fast-paced, hectic lives that involve using them a lot. Hence, balancing one's life requires an amount of hours dedicated toward work, an amount of hours dedicated toward brain-exercising recreation, and an amount of hours towards abolsute rest and brain boredom. For all college students, remember this philosophy, and video games will help you perform the best you can in your academic career.
