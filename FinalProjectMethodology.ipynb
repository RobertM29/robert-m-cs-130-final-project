   - used other_responses_count = TotalResponders_count - (havingfun_count + improving_count + winning_count + relaxing_count + alltheabove_count) to calculate the amount of responses that were most uncommon

   - compared data by finding the amount of respondents with a SWL score higher than 30 (gaming_anxiety.loc[gaming_anxiety.SWL_T > 30]), the amount of respondents with a score higher than 20 (gaming_anxiety.loc[gaming_anxiety.SWL_T > 20], the amount of respondents with a score lower than 20 gaming_anxiety.loc[gaming_anxiety.SWL_T , 20], and the amount of respondents with a score below 10 gaming_anxiety.loc[gaming_anxiety.SWL_T < 10]

   - used highest and lowest score variables to determine the amounts of respondents in these categories, which will convey the range of scores, the min and max
   - used stats_count to find percentages of respondents for each count variable defined above

- SPIN (Social Phobia Inventory) Comparison Methods:

    - Since 15 was the most common score, gaming_anxiety.loc[gaming_anxiety.SPIN_T > 15] and gaming_anxiety.loc[gaming_anxiety.SPIN_T < 15] were used to determine the amount of respondents who had more than the common score, and the amount of respondents who had less than the common score
    - Since less than 19 is considered a good score range, gaming_anxiety.loc[gaming_anxiety.SPIN_T < 19] was used to determine the amount of respondents who do not have extreme anxiety over videogames.
    - count variables (SPINMoreThan60_count, SPINMoreThan15_count, SPINLessThan15_count, SPINLessThan19_count) stored results from the comparison functions defined above, then stats_count determined the percentages per SPIN comparison category defined above

## Assumptions:
 - Work column means how many people in GAD Survey distracted by gaming while they worked 
 - Game column conveys which game among responders was the most played
 - Platform column shows which game genre is most played among responders
 - WhyPlay column conveys the most common reason play videogames, despite their anxiety
 - Degree column conveys the education status of individuals, while work column conveys employment status
 - Work and Degree Columns: compared with the whyplay column to determine if videogames could help with well-being during people's work and college careers. 
 - Birthplace and residence columns can be used to determine which country gives birth to the most videogamers, and which country has the most videogamers residing; comparisons can be made to see if the same country yields these titles, or if two different countries do
 - If a single player plays Clash Royale to relax, then it is likely they want to escape their busy work and social life to spend time alone. However, if a multiplayer person plays Clash Royale to relax, then it is likely they want to escape work and hang out with friends (data can be analyzed to determine people's motives in videogaming, whether it be based on the specific game, or even in general)
 - For Birthplace and Birthplace_ISO3 columns: counts are the same, but lengths are different; this 2 length difference (126 to 124) likely shortens the data a bit (maybe due to abbreviations or maybe 2 counts that were the same for a country - a bunch of them were ones - were combined together to equal 2 twice).
 - All survey respondents have some form of anxiety when gaming. 
 - For SWL (Satisfaction With Life), 35 is the highest score, which is either a perfect score, or a perfect score is 40, which can lead to 2 interpretations: gamers have a little above average (87.5%) peak life satisfaction, or a few gamers have perfect life saitsfaction (100%); it makes the most sense for experimentors to have the highest score be an interval of 5 or 10.
 - Considering work was asked on the survey, the respondents are likely young adults or close to middle aged (20s and 30s).
 - The highest score for SPIN (Social Phobia) is unknown, yet an assumption would be 15 or 20: if 15 is the highest score, then gamers have very high social phobia still (about 93%), however, if the highest score is 20, then their social phobia has dimmed to 70%.
 - Values In Hours Column assumed to be an average per week. If a person inputted 20 as their hour amount, then one can assume they play an average of 20 hours per week.
 - In League Data, Gold and gold are the same rank - only difference is spelling.
 - In Birthplace and Residence Data, inputs are likely all countries. 
 - For Social Phobia (SPIN Data), the highest score is known to be 68 (less than 19 suggests anxiety is ok). 
 - For Work Data, Student at school is assumed to be referring to highschool, and based on the number of respondents who are 18, most of them may still be in highschool, and a few are in college, depending on their birthdate.
 - When comparing work and degree, it can be assumed degree relates to what a college/university student is currently pursuing, or the degree can relate to what a student has already achieved.
