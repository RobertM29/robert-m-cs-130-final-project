```python
def sports_game(): 
    count = 0 
    for element in data['Genre']: 
        if element == 'Sports': 
           count = count + 1  #everytime genre is found count increments to determine the most popular genre (same logic for next 6 functions)
    return count
print(sports_game())
    
    
```
